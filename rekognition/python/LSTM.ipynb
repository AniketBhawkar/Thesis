```python
# LSTM and CNN for sequence classification in the IMDB dataset
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt

from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
import pandas as pd

from sklearn.metrics import mean_squared_error
from math import sqrt

from sklearn.model_selection import train_test_split
```

```python
names = ['L1','L2','L3','L4','L5']
dataset = pd.read_csv("F:/Masters/Semester 3/thesis/rekognition/lstm-dataset-final.csv", delimiter=",",names=names)
dataset = dataset.iloc[1:]
dataset.head()


Data = dataset.iloc[:, :-1].values  
target = dataset.iloc[:, 4].values

```

```python
data = np.array(Data, dtype=float)
target = np.array(target, dtype=float)
```

```python
data.shape
```

```python
target.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, target, test_size=0.25,random_state = 4)

# reshape input to be [samples, time steps, features]
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1],1))
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1],1))

x_test.shape
```

```python
model=Sequential()
```

```python
model.add(LSTM((512),batch_input_shape=(None,None,1),return_sequences=True))
model.add(LSTM((1),return_sequences=False))
```

```python
model.compile(loss='mean_absolute_error',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(x_train,y_train, epochs=500, validation_data=(x_test,y_test))
```

```python
results = model.predict(x_test)
```

```python
plt.scatter(range(29), results, c='r')
plt.scatter(range(29), y_test, c='g')
plt.show()
```

```python
plt.plot(results, c='r')
plt.plot(y_test, c='g')
plt.show()
```

```python
plt.plot(history.history['loss'])

plt.show()
```
